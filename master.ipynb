```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import BertTokenizer, TFBertForSequenceClassification
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
import tensorflow_addons as tfa
import random 
import pandas as pd

# Nahrání dat
data = [ ("Jaký je úrok u spořicího účtu?", "Bankovní produkty a jejich parametry"),
("Můžu si online založit účet?", "Servisní otázky a úkony"),
("Jaké jsou poplatky za vedení účtu?", "Bankovní produkty a jejich parametry"),
("Kde najdu nejbližší pobočku?", "Obecné informace k bance"),
("Kolik stojí výběr z cizího bankomatu?", "Bankovní produkty a jejich parametry"),
("Zapomněl jsem PIN, co s tím?", "Servisní otázky a úkony"),
("Jak změnit limit pro bezkontaktní platby?", "Servisní otázky a úkony"),
("Kde si mohu stáhnout mobilní aplikaci?", "Obecné informace k bance"),
("Mám nárok na studentský účet?", "Bankovní produkty a jejich parametry"),
("Proč mi přišla nižší úroková sazba?", "Bankovní produkty a jejich parametry"),
("Jak aktivovat novou kartu?", "Servisní otázky a úkony"),
("Co je to IBAN?", "Obecné informace k bance"),
("Můžu změnit typ svého účtu online?", "Servisní otázky a úkony"),
("Jaké jsou výhody vaší kreditní karty?", "Bankovní produkty a jejich parametry"),
("Co znamená SEPA platba?", "Obecné informace k bance"),
("Jak zrušit účet?", "Servisní otázky a úkony"),
("Jaké doklady potřebuji k otevření účtu?", "Servisní otázky a úkony"),
("Máte hypotéky pro mladé?", "Bankovní produkty a jejich parametry"),
("Kolik mohu vybrat denně z bankomatu?", "Bankovní produkty a jejich parametry"),
("Co je phishing?", "Obecné informace k bance"),
("Jak funguje přímé zadání?", "Servisní otázky a úkony"),
("Můžete mi poslat novou kartu?", "Servisní otázky a úkony"),
("Jak dlouho trvá převod peněz do zahraničí?", "Bankovní produkty a jejich parametry"),
("Jaké máte půjčky?", "Bankovní produkty a jejich parametry"),
("Jak zablokovat ztracenou kartu?", "Servisní otázky a úkony"),
("Proč je váš web tak pomalý?", "Irelevantní otázky na produkty/služby"),
("Dáte mi slevu na poplatky?", "Irelevantní otázky na produkty/služby"),
("Jaké jsou dnes úrokové sazby?", "Bankovní produkty a jejich parametry"),
("Můžu platit v eurech bez poplatku?", "Bankovní produkty a jejich parametry"),
("Kdy bude hotová moje nová karta?", "Servisní otázky a úkony"),
("Máte účty v cizích měnách?", "Bankovní produkty a jejich parametry"),
("Jak si změním heslo do internetového bankovnictví?", "Servisní otázky a úkony"),
("Je možné zvýšit úvěrový limit?", "Bankovní produkty a jejich parametry"),
("Proč mi byl účet zablokován?", "Servisní otázky a úkony"),
("Máte speciální nabídky pro studenty?", "Bankovní produkty a jejich parametry"),
("Kde nahlásím podezřelou transakci?", "Servisní otázky"),
("Kde najdu nejbližší pobočku?", "Obecné informace k bance"),
("Můžete mi doporučit film?", "Irelevantní otázky na produkty/služby"),
("Jak si mohu změnit limit pro výběr z bankomatu?", "Servisní otázky a úkony"),
("Jaké jsou poplatky za vedení účtu?", "Bankovní produkty a jejich parametry"),
("Máte mobilní aplikaci?", "Obecné informace k bance"),
("Jak dlouho trvá převod peněz do zahraničí?", "Bankovní produkty a jejich parametry"),
("Proč je váš web tak pomalý?", "Nevhodné a hrubé otázky"),
("Jak aktivuji mobilní bankovnictví?", "Servisní otázky a úkony"),
("Co je to IBAN?", "Obecné informace k bance"),
("Jak si mohu objednat novou platební kartu?", "Servisní otázky a úkony"),
("Jaké máte typy úvěrů?", "Bankovní produkty a jejich parametry"),
("Je vaše banka otevřená o víkendu?", "Obecné informace k bance"),
("Proč mi nefunguje internetové bankovnictví?", "Servisní otázky a úkony"),
("Jaký je váš SWIFT kód?", "Obecné informace k bance"),
("Můžete mi poslat nové heslo do bankovnictví?", "Servisní otázky a úkony"),
("Máte nabídku pro studenty?", "Bankovní produkty a jejich parametry"),
("Jak mohu zrušit účet?", "Servisní otázky a úkony"),
("Co znamená vaše logo?", "Obecné informace k bance"),
("Kolik stojí výměna karty?", "Bankovní produkty a jejich parametry"),
("Můžete mi půjčit peníze?", "Irelevantní otázky na produkty/služby"),
("Jak zjistím svůj dluh na kreditní kartě?", "Bankovní produkty a jejich parametry"),
("Jak si mohu změnit adresu?", "Servisní otázky a úkony"),
("Jaké jsou výhody vaší kreditní karty?", "Bankovní produkty a jejich parametry"),
("Jakou máte úrokovou sazbu u hypoték?", "Bankovní produkty a jejich parametry"),
("Můžete mi poradit, jak investovat?", "Servisní otázky a úkony"),
("Kdy budete mít systémovou údržbu?", "Obecné informace k bance"),
("Jste nejlepší banka?", "Irelevantní otázky na produkty/služby"),
("Proč je váš servis tak špatný?", "Nevhodné a hrubé otázky"),
 

("Jste naprostí idioti?", "Nevhodné a hrubé otázky"),
("Proč je váš servis tak hrozně špatný?", "Nevhodné a hrubé otázky"),
("Máte vůbec nějaké vzdělání?", "Nevhodné a hrubé otázky"),
("To jste si vymysleli sami, nebo vám to někdo poradil?", "Nevhodné a hrubé otázky"),
("Je normální, že jste tak neschopní?", "Nevhodné a hrubé otázky"),
("Kdo vás vůbec najal?", "Nevhodné a hrubé otázky"),
("Nemáte pocit, že byste měli raději zavřít?", "Nevhodné a hrubé otázky"),
("Tak málo inteligence jsem už dlouho neviděl.", "Nevhodné a hrubé otázky"),
("Myslíte, že tohle je přijatelné chování?", "Nevhodné a hrubé otázky"),
("Je váš manažer taky takový idiot?", "Nevhodné a hrubé otázky"),
("Tohle je důvod, proč od vás lidé odcházejí.", "Nevhodné a hrubé otázky"),
("Můžete mi vysvětlit, proč jste tak neslušní?", "Nevhodné a hrubé otázky"),
("Vážně si myslíte, že se takhle máte chovat k zákazníkům?", "Nevhodné a hrubé otázky"),
("Máte v plánu se někdy zlepšit, nebo tohle je váš vrchol?", "Nevhodné a hrubé otázky"),
("Proč se chováte jako poslední zmetci?", "Nevhodné a hrubé otázky"),
("Nemáte pocit, že byste měli vrátit peníze za tak špatnou službu?", "Nevhodné a hrubé otázky"),
("Jak je možné, že jste ještě nezkrachovali?", "Nevhodné a hrubé otázky"),
("Je tohle váš způsob, jak odradit klienty?", "Nevhodné a hrubé otázky"),
("Myslíte, že je normální ignorovat zákazníky?", "Nevhodné a hrubé otázky"),
("Jak dlouho vás ještě budou trpět na této pozici?", "Nevhodné a hrubé otázky"),
("Jak mohu zkontrolovat zůstatek na svém účtu?", "Bankovní produkty a jejich parametry"),
("Jaké jsou aktuální úrokové sazby pro spořicí účty?", "Bankovní produkty a jejich parametry"),
("Můžu založit účet bez návštěvy pobočky?", "Servisní otázky a úkony"),
("Jak si mohu změnit PIN k mé kartě?", "Servisní otázky a úkony"),
("Kde naleznu informace o vašich nových produktech?", "Obecné informace k bance"),
("Jaké jsou výhody internetového bankovnictví?", "Obecné informace k bance"),
("Proč mi byl účet zablokován bez varování?", "Nevhodné a hrubé otázky"),
("Je vaše zákaznická podpora vždy takto nekompetentní?", "Nevhodné a hrubé otázky"),
("Můžete mi ukázat, jak otevřít spořicí účet online?", "Servisní otázky a úkony"),
("Jak dlouho trvá schválení půjčky?", "Bankovní produkty a jejich parametry"),
("Je možné převádět peníze do zahraničí pomocí vaší aplikace?", "Servisní otázky a úkony"),
("Máte nějaké bezpečnostní tipy pro používání platebních karet?", "Obecné informace k bance"),
("Proč je vaše mobilní aplikace tak pomalá?", "Nevhodné a hrubé otázky"),
("Jaké dokumenty potřebuji pro otevření firemního účtu?", "Bankovní produkty a jejich parametry"),
("Můžete mi poradit, jak lépe spravovat můj účet?", "Servisní otázky a úkony"),
("Jak se mohu přihlásit do internetového bankovnictví?", "Servisní otázky a úkony"),
("Jaké jsou poplatky za mezinárodní převody?", "Bankovní produkty a jejich parametry"),
("Můžete mi doporučit vhodný spořicí účet?", "Bankovní produkty a jejich parametry"),
("Jak zruším svůj účet u vaší banky?", "Servisní otázky a úkony"),
("Proč moje karta nefunguje, když mám na účtu dostatek peněz?", "Servisní otázky a úkony"),
("Máte nějaké speciální nabídky pro studenty?", "Bankovní produkty a jejich parametry"),
("Jak mohu získat kreditní kartu?", "Bankovní produkty a jejich parametry"),
("Proč máte tak vysoké poplatky za vedení účtu?", "Nevhodné a hrubé otázky"),
("Nemůžete udělat nic pro to, abyste mi pomohli rychleji?", "Nevhodné a hrubé otázky"),
("Jak mohu aktivovat svou novou kartu?", "Servisní otázky a úkony"),
("Kde najdu seznam všech vašich poboček?", "Obecné informace k bance"),
("Můžete mi vysvětlit, proč jsem dostal poplatek za toto?", "Servisní otázky a úkony"),
("Vaše služby jsou horší než konkurence, proč bych měl zůstat?", "Nevhodné a hrubé otázky"),
("Jak dlouho mám čekat na odpověď od zákaznické podpory?", "Servisní otázky a úkony"),
("Je možné mít více debetních karet k jednomu účtu?", "Bankovní produkty a jejich parametry"),
("Můžete mi pomoci najít můj IBAN?", "Servisní otázky a úkony"),
("Proč je tak těžké se dostat do vašeho internetového bankovnictví?", "Nevhodné a hrubé otázky"),
("Máte nějaké tipy, jak si udržet dobrý kreditní skóre?", "Obecné informace k bance"),
("Jak mohu oznámit podezřelou transakci?", "Servisní otázky a úkony"),
("Proč je váš web neustále nedostupný?", "Nevhodné a hrubé otázky"),
("Jaké jsou podmínky pro získání hypotéky?", "Bankovní produkty a jejich parametry"),
("Kolik mohu vybrat z ATM najednou?", "Bankovní produkty a jejich parametry"),
("Jak se mohu stát vaším klientem?", "Servisní otázky a úkony"),
("Jaké jsou výhody vaší premium kreditní karty?", "Bankovní produkty a jejich parametry"),
("Co dělat, když jsem zapomněl své heslo do internetového bankovnictví?", "Servisní otázky a úkony"),
("Jaké jsou poplatky za zrušení účtu?", "Bankovní produkty a jejich parametry"),
("Kolik stojí mezinárodní převod peněz?", "Bankovní produkty a jejich parametry"),
("Je bezpečné používat vaši mobilní aplikaci na veřejném Wi-Fi?", "Obecné informace k bance"),
("Jak mohu upgradovat svůj účet?", "Servisní otázky a úkony"),
("Co znamená vaše logo?", "Irelevantní otázky na produkty/služby"),
("Můžete mi doporučit nějaké finanční poradenství?", "Servisní otázky"),
("Jak si mohu založit spořicí účet?", "Bankovní produkty a jejich parametry"),
("Proč jste nejlepší banka?", "Irelevantní otázky na produkty/služby"),
("Můžu si u vás zřídit účet i když nejsem občanem?", "Servisní otázky a úkony")
 ]


# random.shuffle(data)

# print(data[:185])
# df = pd.DataFrame(data, columns=['Question', 'Category'])

# pd.set_option('display.max_rows', None)

# print(df)

# Tímto způsobem rozděluji data na otázky a kategorie
questions, categories = zip(*data)

# Tyto otázky a kategorie je třeba encodovat do čísel
label_encoder = LabelEncoder()
categories_encoded = label_encoder.fit_transform(categories)
num_classes = np.max(categories_encoded) + 1

# Klasické rozdělení dat na testovací a trenovací data
X_train, X_test, y_train, y_test = train_test_split(questions, categories_encoded, test_size=0.3, random_state=42)

# Načtení transformeru Berta
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Zde nastavuji data do formátu, který zpracuje Bert. Bylo třeba zde oříznout data na maximální délku 128, 
# aby pracoval správně. V tomto nastavení padding=True nastavuje pravidlo, aby byly všechna data stejně dlouhá. 
# # Pokud není, tak jsou doplněná o [PAD] tokeny tzn. například o číslo 0
train_encodings = tokenizer(list(X_train), truncation=True, padding=True, max_length=128)
test_encodings = tokenizer(list(X_test), truncation=True, padding=True, max_length=128)

# V této sekci řeším rozpad otázek na tokeny, aby je mohl zpracovat Bert. Funkce dict funguje jako formát slovníku, 
# který rozklíčuje názvy vstupních vrstev modelu. Např input_ids, attention_mas atp.
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))

# Zde načítám předtrénovaný model Bert
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_classes)

optimizer = Adam(learning_rate=5e-5) 
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True) 
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy']) 

# Trénink modelu
model.fit(train_dataset.shuffle(len(X_train)).batch(32), epochs=20, batch_size=32, validation_data=test_dataset.batch(60))
model.evaluate(test_dataset.batch(32))
```

```python
# Nastavení funkce, která odpoví na danou otázku dle kategorie do které spadá díky predikci modelu
def classify_and_respond(text):
    inputs = tokenizer(text, return_tensors="tf", truncation=True, padding=True, max_length=128)
    predictions = model(inputs.data)[0]
    # Zde získávám index nejvyšší predikované předpovědi. Tzn. Identifikace kategorie, kterou model predikuje jako 
    # nejpravděpodobnější pro daný vstup a to zapomoci nejvyšší pravděpodobnosti z modelu. Index 0 znamená, 
    # že chceme pracovat pouze s předpovědmi, které se nacháyí na tomto indexu a ne s dalšímz
    predicted_class_id = tf.argmax(predictions, axis=1).numpy()[0]
    # Převod predikovaného indexu zpět na kategorii
    predicted_category = label_encoder.inverse_transform([predicted_class_id])[0]

    responses = {
        "Bankovní produkty a jejich parametry": "Informace o bankovních produktech najdete na našem webu.",
        "Servisní otázky a úkony": "Pro servisní otázky nás prosím kontaktujte přes náš zákaznický servis.",
        "Servisní otázky": "Pro dotazy týkající se našich služeb a vašich možností, neváhejte vyhledat podrobnější informace na naší webové stránce nebo se obrátit na naši zákaznickou podporu.",
        "Obecné informace k bance": "Obecné informace o bance naleznete na našich webových stránkách.",
        "Irelevantní otázky na produkty/služby": "Omlouváme se, ale nemůžeme odpovědět na irelevantní dotazy.",
        "Nevhodné a hrubé otázky": "Prosíme, abyste zachovali slušné chování. Máte-li nějaké konkrétní dotazy, rádi vám pomůžeme.",
    }
    
    response = responses.get(predicted_category, "Omlouváme se, ale na tuto otázku nemůžeme odpovědět.")
    print(f"Dotaz: {text}\nKategorie: {predicted_category}\nOdpověď: {response}")


test_question = "Jsi úplně k ničemu, víš o tom?"
classify_and_respond(test_question)
```

```python
from sklearn.metrics import classification_report
import numpy as np

y_pred = model.predict(test_dataset.batch(32))
y_pred = np.argmax(y_pred.logits, axis=1)

y_true = np.array(y_test)

labels = np.unique(np.concatenate([y_true, y_pred]))
target_names = label_encoder.inverse_transform(labels)

print(classification_report(y_true, y_pred, labels=labels, target_names=target_names))
```
